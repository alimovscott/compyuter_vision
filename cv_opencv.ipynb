```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat.jpeg")
cv2_imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized =cv2.resize(gray,(150,100))
cv2_imshow(resized)
```

```python
cropped = gray[100:300, 150:350]
cv2_imshow(cropped)
```

```python
rotated =cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray,1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped_cat.jpg", flipped)
```

```python
# video detector
print("Hello, World!")

import cv2

cap = cv2.VideoCapture(0) # kamerani ochadi
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') # yuzni aniqlash uchun Haar Cascade modelini yuklaydi

while True:
    success, frame = cap.read() # True yoki false qaytaradi, frame esa rasmni o'z ichiga oladi
    if not success:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # rasmni kulrang rangga o'zgartiradi

    faces = face_cascade.detectMultiScale(gray, 1.1, 4,) # yuzlarni aniqlash uchun detectMultiScale metodini chaqiradi

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2) # yuzlarni to'rtburchak bilan o'rab chiqadi

    cv2.imshow("Frame", frame) # rasmni oynada ko'rsatadi

    if cv2.waitKey(1) & 0xFF == ord('q'): # 'q' tugmasi bosilganda dasturdan chiqadi
        break

cap.release() # kamerani yopadi
cv2.destroyAllWindows() # barcha oynalarni yopadi

##########################################

# import cv2


# url = "rtsp://admin:admin@192.168.1.100:554/stream1" # streaming URL manzili

# cap = cv2.VideoCapture(0)


# while True:
#     success, frame = cap.read() # True yoki false qaytaradi, frame esa rasmni o'z ichiga oladi
#     if not success:
#         break


#     gray= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # rasmni kulrang rangga o'zgartiradi
#     blurred = cv2.GaussianBlur(frame, (15,15), 0) # rasmni burish uchun Gaussian blur qo'llaydi
#     edges = cv2.Canny(frame, 100, 200) # rasmning chekkalarini aniqlash uchun Canny algoritmini qo'llaydi
#     flipped = cv2.flip(frame, 1) # rasmni gorizontal ravishda aylantiradi

#     cv2.imshow("Orginal", frame) # rasmni oynada ko'rsatadi
#     # cv2.imshow("Gray", gray) # kulrang rasmni oynada ko'rsatadi
#     # cv2.imshow("Edges", edges) # chekkalarni ko'rsatadi
#     # cv2.imshow("Blurred", blurred) # burish rasmni oynada ko'rsatadi noise
#     cv2.imshow("Flipped", flipped) # aylantirilgan rasmni oynada ko'rsatadi




#     if cv2.waitKey(1) & 0xFF == ord('q'): # 'q' tugmasi bosilganda dasturdan chiqadi
#         break


# cap.release() # kamerani yopadi
# cv2.destroyAllWindows() # barcha oynalarni yopadi

```
