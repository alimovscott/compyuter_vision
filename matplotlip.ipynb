```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("y Values")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])


plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue',linestyle='--', marker='o')
plt.title('Multiple Lines Example')
plt.xlabel('X Values')
plt.xlabel('Y values')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['Salary', 'Expense', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y, color='purple', s=100)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])

y1 = np.array([10, 15, 20, 25])
y2 = np.array([8, 12, 18, 22])

plt.plot(x, y1, label='Dataset A', color='red', marker='o')
plt.plot(x, y2, label='Dataset B', color='blue', marker='s', linestyle='--')

plt.title("Matplotlib with NumPy Arrays")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.legend()
plt.grid(True)

plt.show()
```
